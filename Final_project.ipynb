Simple linear regression
at this point i assume that the rate of use of cigarettes is independent to the tax rate, meaning that the dependent variable, tax is equal to zero

```
attach(x)
model<-lm(Cig.use~Tax)
```

REGRESSION LINE

```
plot(Tax,Cig.use, pch=21, cex=0.5, bg='green', main="Cigrate Use VS Tax with Regression Line")
abline(model$coef, lwd=2)
```

```
res<- model$ residuals;
 plot(Tax,res,main="Residuals") 
```

**3)Results**
The coefficient of Tax is -1.8275, which means when tax rate increase 1 unit which is 1 percent, the cigarette use rate will decrease -1.8275%. And its p-value is 0.000201. Thus the coefficient of Tax is significant at 99% level.

The intercept is 22.0126 means when tax rate equals zero, the cigarette use rate is 22.0126%.

R-square is 0.2524 which means tax rate explained 25.24% of cigarette use rate. The residuals for the regression behave as a normal distribution with mean zero.

Thus, this regression is a fitted model. But to declare the true relationship between cigarette use rate and tax rate should be further determined.
